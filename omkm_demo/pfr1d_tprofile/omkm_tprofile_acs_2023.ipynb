## 2. TProfiles used in this example

### 2.1 See the Plot below


```python
import pandas as pd
import seaborn as sns


pfr_const = pd.read_csv("constant/rctr_state_ss.csv")
pfr_inc = pd.read_csv("increasing/rctr_state_ss.csv")
pfr_para = pd.read_csv("parabola/rctr_state_ss.csv")

sns.lineplot(data=pfr_inc,
                x="z(m)",
                y="Temperature",
                label="Increasing T-Profile",
                marker='s'
             )
sns.lineplot(data=pfr_const,
                x="z(m)",
                y="Temperature",
                label="Constant T-Profile",
                marker='o'
             )
sns.lineplot(data=pfr_para,
                x="z(m)",
                y="Temperature",
                label="Parabolic T-Profile",
                marker='<'
             )
```

## 4. Results of the simulation

### 4.1 Mass fraction of Ethylene (C$_2$H$_4$)

```python
pfr_const = pd.read_csv("constant/gas_mass_ss.csv")
pfr_inc = pd.read_csv("increasing/gas_mass_ss.csv")
pfr_para = pd.read_csv("parabola/gas_mass_ss.csv")

name = "CH2CH2"
sns.lineplot(data=pfr_inc,
                x="z(m)",
                y=name,
                label="Increasing T-Profile",
                marker='s'
             )
sns.lineplot(data=pfr_const,
                x="z(m)",
                y=name,
                label="Constant T-Profile",
                marker='o'
             )
sns.lineplot(data=pfr_para,
                x="z(m)",
                y=name,
                label="Parabolic T-Profile",
                marker='<'
             )
```

### 4.3 Fractional coverage of C(S)

```python
pfr_const = pd.read_csv("constant/surf_cov_ss.csv")
pfr_inc = pd.read_csv("increasing/surf_cov_ss.csv")
pfr_para = pd.read_csv("parabola/surf_cov_ss.csv")

name = "C(S)"
sns.lineplot(data=pfr_inc,
                x="z(m)",
                y=name,
                label="Increasing T-Profile",
                marker='s'
             )
sns.lineplot(data=pfr_const,
                x="z(m)",
                y=name,
                label="Constant T-Profile",
                marker='o'
             )
sns.lineplot(data=pfr_para,
                x="z(m)",
                y=name,
                label="Parabolic T-Profile",
                marker='<'
             )
```
